```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 4.0, 100)
```

```python
def create_sinusoidal(A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

***Creating a Sinusoidal Signal and 3 different plots in subplots***

```python
x = create_sinusoidal(1,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)

ax[0, 0].plot(t,x)
ax[0, 0].set_title('continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('scatter plot', fontsize = 15)
ax[-1,-1].axis('off')


plt.show()
```

***Creating 3 Sinusoidal signals and then combine and plot the signal***

```python
x1 = create_sinusoidal(1,4)
x2 = create_sinusoidal(2,5)
x3 = create_sinusoidal(4,7)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal signal and a combined signal', fontsize=15)

ax[0, 0].plot(t,x1, 'b')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'r')
ax[1,0].set_title('x3(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3)
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```
