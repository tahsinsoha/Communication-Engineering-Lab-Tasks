```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,2.0,0.00001)
```

```python
def create_sinusoidal(t, A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi* f * t + phase)
  return x
```

```python
x1 = create_sinusoidal(t, 1, 4)
x2 = create_sinusoidal(t, 1, 11)
x3 = create_sinusoidal(t, 1, 16)
y = x1 + x2 + x3
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=15)
ax[0,0].plot(t, x1)
ax[0,0].set_title('Signal-1', fontsize = 15)

ax[0,1].plot(t, x2)
ax[0,1].set_title('Signal-2', fontsize = 15)

ax[1,0].plot(t, x3)
ax[1,0].set_title('Signal-3', fontsize = 15)

ax[1,1].plot(t, y)
ax[1,1].set_title('Signal-1 + Signal-2 + Signal-3', fontsize = 15)

plt.show()
```

```python
fs = 12
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 4) + create_sinusoidal(n, 1, 8) + create_sinusoidal(n, 1, 12)
plt.title('fs = 12', fontsize = 15)
plt.stem(n,y)
plt.show()

fs = 24
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs = 24', fontsize = 15)
plt.stem(n,y)
plt.show()

fs = 36
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs = 36', fontsize = 15)
plt.stem(n,y)

plt.show()
```
