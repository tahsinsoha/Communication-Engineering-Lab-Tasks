```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Function for plotting signal
def plot_signal(t, x):
  plt.figure(figsize=(12, 6))
  plt.plot(t, x)
  plt.show()
```


***Amplitude Modulation(AM) ***

***Step 1: Creating Carrier Signal***

```python
Ac=2; 
fc=0.5; 
t=np.arange(0, 50, 0.1)
cs=Ac*np.sin(2*np.pi*fc*t); 
plot_signal(t, cs)
```

***Step 2: Creating message Signal***

```python
ms = 1.0
M = Ac * ms
fm=.05
phase=0
mt=M*np.cos((2*np.pi*fm*t)+phase)
plot_signal(t, mt)
```

***Step 3: Computing Amplitude modulation***

```python
y=(1+ms*np.cos((2*np.pi*fm*t) + phase))*Ac*np.sin(2*np.pi*fc*t);
plot_signal(t, y)
```

***Frequency Modulation(FM)***

***Step 1: Creating Carrier Signal***


```python
Ac=2; 
fc=40; 
t=np.arange(0, 1.0, 0.0001)
cs=Ac*np.cos(2*np.pi*fc*t); 
plot_signal(t, cs)
```

***Step 2: Creating modulating signal***

```python
m = 1.
fm=4
ms=np.sin((2*np.pi*fm*t))
plot_signal(t, ms)
```

***Step 3: Computing Frequency modulation***


```python
y = Ac*np.cos(2. * np.pi * (fc * t + m * np.sin(2*np.pi*fm*t)))
plot_signal(t, y)
```


***Phase Modulation(PM) ***

***Step 1: Creating Carrier Signal***


```python
Ac=2; 
fc=2; 
t=np.arange(0, 4.0, 0.0005)
phase = 0
cs=Ac*np.sin(2*np.pi*fc*t + phase); 
plot_signal(t, cs)
```

***Step 2: Creating message Signal***

```python
k = 1.0
fm=4
Am = 2
ms=Am*np.sin(2 * np.pi*fm*t)
plot_signal(t, ms)
```

***Step 3: Computing Phase modulation***

```python
y=Ac*np.sin((2*np.pi*fc*t) + (k * ms + phase))
plot_signal(t, y)
```
